```python
# Given an array of integers and a target value, return the indices of the two numbers whose sum equals the target.
# nums = [2, 7, 11, 15]
# target = 9

# A Hash Map is a data structure that stores data as key-value pairs.
#  It uses a hash function to map each key to a memory location, allowing data to be inserted, searched, and updated in O(1) average time.


def two_sum(nums, target):
    seen = {}
    for i in range(len(nums)):
        rem = target - nums[i]
        
        if rem in seen:
            return [seen[rem],i]
        
        else:
            seen[nums[i]] = i
            
    return []
    
nums = [2, 7, 11, 15]
target = 18

t = two_sum(nums,target)

print(t)
        
```

```python
# 2. Best Time to Buy and Sell Stock ⭐⭐⭐⭐



def max_profit(prices):
    const = 0
    for i in range(len(prices)):
        for j in range(i+1,len(prices)):
            amt = prices[j] - prices[i]
            if amt > const:
                const = amt
    return const
                
                
prices = [7,1,5,3,6,4]

a = max_profit(prices)

print(a)
        
```
